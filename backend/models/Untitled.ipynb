```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the dataset
# data = pd.read_csv("transactions.csv")

# Extract the features and target variables
X = data[['Amount', 'Transaction Type', 'Category', 'Account Name']]
y_category = data['Category']
y_price = data['Amount']

# Encode the categorical features
X = pd.get_dummies(X, columns=['Transaction Type', 'Category', 'Account Name'])

# Split the dataset into training and testing sets
X_train, X_test, y_train_category, y_test_category, y_train_price, y_test_price = train_test_split(X, y_category, y_price, test_size=0.3, random_state=42)

# Train the category prediction model using RandomForestClassifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train_category)

# Train the price prediction model using LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train_price)

from sklearn.metrics import classification_report, accuracy_score

# Make predictions on the test set
y_pred_category = clf.predict(X_test)

# Evaluate the accuracy of the category prediction model
print(f"Accuracy score: {accuracy_score(y_test_category, y_pred_category)}")
print("Classification report:")
print(classification_report(y_test_category, y_pred_category))

from sklearn.metrics import r2_score

# Make predictions on the test set
y_pred_price = reg.predict(X_test)

# Evaluate the accuracy of the price prediction model
print(f"R-squared score: {r2_score(y_test_price, y_pred_price)}")
```

```python
import pandas as pd

# Load the original dataset
# df = pd.read_csv('dataset.csv')

# Create a new dataframe to store the previous 30 entries for each row
new_df = pd.DataFrame()

# Loop over each row in the original dataframe
for i in range(len(df)):
    # Get the date of the current row
    current_date = pd.to_datetime(df.loc[i, 'Date'], format='%m/%d/%Y')
    
    # Get the previous 30 dates
    prev_dates = pd.date_range(current_date - pd.Timedelta(days=30), current_date - pd.Timedelta(days=1), freq='D')
    
    # Filter the original dataframe to get the previous 30 entries
    prev_entries = df[(pd.to_datetime(df['Date'], format='%m/%d/%Y').isin(prev_dates)) & (df['Account Name'] == df.loc[i, 'Account Name'])]
    
    # Pivot the previous entries to get the amounts by category
    pivot_table = prev_entries.pivot_table(values='Amount', index='Account Name', columns='Category', aggfunc='sum')
    
    # Add the pivot table to the new dataframe
    new_df = new_df.append(pivot_table, ignore_index=True)

# Add the columns for the account name and date to the new dataframe
new_df.insert(0, 'Account Name', df['Account Name'])
new_df.insert(1, 'Date', df['Date'])

# Fill any missing values with 0
new_df = new_df.fillna(0)
```
